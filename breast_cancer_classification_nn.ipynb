### Breast Cancer Classification using Neural Network in Python.


Tensorflow and Keras are used to build a simple neural network.
Here's the link to the dataset for reference: https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data?resource=download

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
df1 = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
df1.head()
```

```python
# adding the 'target' column to the dataframe
df1['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
df1.tail()
```

```python
# number of rows and columns in the dataset
df1.shape
```

```python
# getting some information about the data
df1.info()
```

```python
# checking for null values
df1.isnull().sum()
```

```python
# statistical measures about the data
df1.describe()
```

```python
# checking the distribution of target variable
df1['label'].value_counts()
```

1: Benign
0: Malignant

```python
df1.groupby('label').mean()
```

```python
X = df1.drop(columns='label', axis=1)
Y = df1['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

**Building the Neural Network**

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation = 'relu'),
    keras.layers.Dense(2, activation = 'sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

model.predict() gives the prediction probability of each class for that data point.

```python
# converting the prediction probability to class labels

Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

#### Building the Predictive System 

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input_data to a numpy array
input_data_numpy_arr = np.asarray(input_data)

# reshape the numpy array as we are predicting for 1 data point
input_data_reshaped = input_data_numpy_arr.reshape(1, -1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0] == 0):
    print('The tumor is malignant.')

else:
    print('The tumor is benign')
```
